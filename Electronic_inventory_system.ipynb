# purchased

```python
purchased_id =input("enter product id:")
purchased_quantity = int(input("enter  quantity:"))


print("product name",inventory[purchased_id]['prod_name'])
print("price ",inventory[purchased_id]['price'])

print("----------------------------------")

print("Billing amount:",inventory[purchased_id]['price']*purchased_quantity)

print("-----------------------------------")


inventory[purchased_id]['quantity'] = inventory[purchased_id]['quantity'] - purchased_quantity

print("***********************************")
```

```python
var=json.dumps(inventory)
fd= open('records.json','w')
fd.write(var)
fd.close()
```
